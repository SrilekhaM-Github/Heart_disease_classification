```python
x=ab.iloc[:, :-1]
x
```

```python
y=ab.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)
```

```python
fig,ax=plt.subplots(1,1,figsize=(18,12))
sns.heatmap(ab.corr(),annot=True,linewidths=1)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier
c=AdaBoostClassifier()
c.fit(x,y)
d=c.predict(xtest)
e=accuracy_score(ytest,d)
print(e)
```

```python
from sklearn.neighbors import KNeighborsClassifier
c=KNeighborsClassifier()
c.fit(x,y)
d=c.predict(xtest)
e=accuracy_score(ytest,d)
print(e)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

c=GradientBoostingClassifier()
c.fit(x,y)
d=c.predict(xtest)
e=accuracy_score(ytest,d)
print(e)
```

```python
prediction=c.predict([[63,1,3,145,233,1,0,150,0,2.3,0,0,1]])
print(prediction)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,d))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,d)
cm
```
